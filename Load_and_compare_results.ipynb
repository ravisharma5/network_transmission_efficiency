```python
llm1_name="codellama-7b-instruct.Q5_K_M"
llm2_name="granite-8b-code-instruct"
import numpy as np
```

```python
# Load results and plot them
def load_results(filename):
    """
    Load the results from a saved .npz file.

    Parameters:
        filename (str): Path to the file.

    Returns:
        dict: Loaded results.
    """
    data = np.load(filename)
    results = {
        'batch_sizes': data['batch_sizes'],
        'OPT_SE': data['OPT_SE'], 'OPT_EE': data['OPT_EE'],
        'PROP_SE': data['PROP_SE'], 'PROP_EE': data['PROP_EE'],
        'RAND_SE': data['RAND_SE'], 'RAND_EE': data['RAND_EE'],
        'BIN_SE': data['BIN_SE'], 'BIN_EE': data['BIN_EE']
    }
    return results
```

```python
# Load results for two LLMs
llm1_results = load_results("simulation_results_"+llm1_name+".npz")
llm2_results = load_results("simulation_results_"+llm2_name+".npz")
```

```python
# Plotting OPT_SE comparison between two LLMs
import matplotlib.pyplot as plt

batch_sizes = llm1_results['batch_sizes']

plt.figure(figsize=(10, 6))
plt.plot(batch_sizes, llm1_results['OPT_SE'], label=llm1_name+"_OPT_SE", marker='o')
plt.plot(batch_sizes, llm2_results['OPT_SE'], label=llm2_name+"_OPT_SE", marker='x')

plt.xlabel('Batch Size')
plt.ylabel('OPT_SE')
plt.title('OPT_SE Comparison Between LLM1 and LLM2')
plt.legend()
plt.grid(True)
plt.show()
```
